Capturing video[multiple images ]    To exit key q is used

```python
vid = cv2.VideoCapture(0)
while True:  ## infinite capture from camera
    flag,img = vid.read()
    if flag:
        cv2.imshow('preview',img)
        key=cv2.waitKey(1)
        if key == ord('q'):   ## key to exit the priview    ,, ord('') -- gives ASCII value of the keys or we can also replace it by its ASCII value
            break
    else:
        break  ## if not reading any input from camera
vid.release() ## to release camera access by code
cv2.destroyAllWindows() 
cv2.waitKey(1)
```

IMAGE IN GRAY BLACK AND WHITE

```python
vid = cv2.VideoCapture(0)
while True:  ## infinite capture from camera
    flag,img = vid.read()
    if flag:
        ##Processing code
        img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)  
        th,img_bw = cv2.threshold(img_gray,100,255,cv2.THRESH_BINARY_INV)    ## for inverted video  ,, gray , black and white
        cv2.imshow('preview',img_bw)
        key=cv2.waitKey(1)
        if key == ord('q'):   ## key to exit the priview    ,, ord('') -- gives ASCII value of the keys or we can also replace it by its ASCII value
            break
    else:
        print('NO FRAME')
        break  ## if not reading any input from camera
cv2.destroyAllWindows() 
cv2.waitKey(1)
vid.release() ## to release camera access by code
```
